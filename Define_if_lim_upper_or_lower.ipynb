```python
from __future__ import print_function
from no_big_db_func import get_all_names
from datahandling import access_db
from tinydb import Query
Q = Query()
```

```python
names = get_all_names()
```

```python
mod_db = access_db(3, True)
```

```python
for name in names:
    equip, d_type = name.split(' ')
    
    my_Q = ((Q.equipment_name == equip) & (Q.data_type == d_type))
    
    entry = mod_db.search(my_Q)

    if 'desired_vals' in entry[0]:
        print(name, 'done')
        print('-------------------------')
        continue
    
    done = False
    print(name)
    while not done:
        high_or_low = input("Are the desired values the high or low values: (enter 0 for high and 1 for low)")
    
        if high_or_low not in [0, 1]:
            print('INVALID ENTRY')
            continue
            
        done = True
        
    if high_or_low == 0:
        high_or_low = 'high'
    elif high_or_low == 1:
        high_or_low = 'low'
        
    mod_db.update({'desired_vals': high_or_low}, my_Q)
        
    print('-------------------------')
```
